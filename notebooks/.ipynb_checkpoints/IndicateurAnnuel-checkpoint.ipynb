```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/home/jovyan/work/indicateurs-annuels-cvs-mars22.xlsx'

df = pd.read_excel(file_path, header=7)

df.head()
```

##  Nettoyage des données: Remplacement des valeurs NaN par 0

```python
df.fillna(0, inplace=True)

for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(' ', ''), errors='coerce')
```

```python
sns.set(style="darkgrid")

def plot_trend(df, row_index):
    plt.figure(figsize=(12, 6)) 
    sns.lineplot(x=df.columns[1:], y=df.iloc[row_index, 1:], marker='o', color='orange') 
    plt.title(df.iloc[row_index, 0], fontsize=14)  
    plt.xlabel('Année', fontsize=12)  
    plt.ylabel('Nombre', fontsize=12)  
    plt.xticks(rotation=45, fontsize=10) 
    plt.yticks(fontsize=10)  
    plt.tight_layout() 
    plt.show()
```

```python
# Cambriolages et tentatives de cambriolage
plot_trend(df, 0)

# Vols sans effraction
plot_trend(df, 1)

# Actes de vandalisme contre le logement
plot_trend(df, 2)

# Vols et tentatives de vols de voiture
plot_trend(df, 3)

# Vols et tentatives de vols de deux-roues à moteur
plot_trend(df, 4)

# Vols et tentatives de vol de vélo
plot_trend(df, 5)

# Vols d'objet dans ou sur la voiture
plot_trend(df, 6)

# Actes de vandalisme contre la voiture
plot_trend(df, 7)

# Victimes de vols ou tentatives de vols sans violences ni menaces
plot_trend(df, 8)

# Victimes de vols ou tentatives de vols avec violences ou menaces
plot_trend(df, 9)

plot_trend(df, 10)

plot_trend(df, 11)

plot_trend(df, 12)

plot_trend(df, 13)

plot_trend(df, 14)

plot_trend(df, 15)

plot_trend(df, 16)

plot_trend(df, 17)

plot_trend(df, 18)


```

```python
plt.figure(figsize=(12, 6))
for row_index in [0, 1, 3]:
    plt.plot(df.columns[1:], df.iloc[row_index, 1:], marker='o', label=df.iloc[row_index, 0])
plt.xlabel('Année')
plt.ylabel('Nombre')
plt.title('Comparaison des Tendances de Différents Indicateurs')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

# Conclusion : 

## Y a-t-il une tendance globale à l'augmentation ou à la diminution de la criminalité au fil des années ?
En ce qui concerne les cambriolages et les tentatives de cambriolage visant les résidences principales, il y a une diminution significative depuis 2006.
Pour les vols sans effraction visant les résidences principales, il semble y avoir des fluctuations, mais la tendance n'est pas claire.
Les actes de vandalisme contre le logement ont connu des variations, mais ils semblent diminuer récemment.
Le nombre de vols de véhicules a diminué depuis 2006, mais il y a des variations d'année en année.
## Qu'en est-il du sentiment d'insécurité ?
Le nombre de personnes se sentant souvent ou parfois en insécurité dans leur quartier ou village semble augmenter régulièrement depuis 2006.
En revanche, le nombre de personnes se sentant souvent ou parfois en insécurité à leur domicile a diminué depuis 2007, avec quelques fluctuations.
## Qu'en est-il des violences physiques et sexuelles ?
Le nombre de victimes de violences physiques hors ménage semble diminuer depuis 2007.
Le nombre de victimes de violences sexuelles hors ménage varie d'année en année sans tendance claire.
Le nombre de victimes de violences physiques et/ou sexuelles au sein du ménage semble diminuer récemment.
## Y a-t-il des ruptures de série ou des données manquantes à prendre en compte ?
Certaines données sont marquées comme "nd" (non disponibles) ou "r" (rupture de série) pour certaines années, ce qui peut indiquer des lacunes ou des changements méthodologiques dans la collecte de données.
