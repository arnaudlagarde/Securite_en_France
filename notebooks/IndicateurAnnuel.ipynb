```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/home/jovyan/work/indicateurs-annuels-cvs-mars22.xlsx'

df = pd.read_excel(file_path, header=7)

df.head()
```

##  Nettoyage des données: Remplacement des valeurs NaN par 0

```python
df.fillna(0, inplace=True)

for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(' ', ''), errors='coerce')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuration du style de graphique
sns.set(style="darkgrid")

def plot_trend(df, row_index):
    plt.figure(figsize=(12, 6))  # Ajustement de la taille du graphique pour une meilleure visibilité
    sns.lineplot(x=df.columns[1:], y=df.iloc[row_index, 1:], marker='o', color='orange')  # Utilisation de seaborn pour une meilleure esthétique
    plt.title(df.iloc[row_index, 0], fontsize=14)  # Augmentation de la taille du titre
    plt.xlabel('Année', fontsize=12)  # Augmentation de la taille de l'étiquette de l'axe x
    plt.ylabel('Nombre', fontsize=12)  # Augmentation de la taille de l'étiquette de l'axe y
    plt.xticks(rotation=45, fontsize=10)  # Rotation et taille de police des étiquettes de l'axe x
    plt.yticks(fontsize=10)  # Taille de police des étiquettes de l'axe y
    plt.tight_layout()  # Optimisation de la mise en page
    plt.show()
```

```python
# Cambriolages et tentatives de cambriolage
plot_trend(df, 0)

# Vols sans effraction
plot_trend(df, 1)

# Actes de vandalisme contre le logement
plot_trend(df, 2)

# Vols et tentatives de vols de voiture
plot_trend(df, 3)

# Vols et tentatives de vols de deux-roues à moteur
plot_trend(df, 4)

# Vols et tentatives de vol de vélo
plot_trend(df, 5)

# Vols d'objet dans ou sur la voiture
plot_trend(df, 6)

# Actes de vandalisme contre la voiture
plot_trend(df, 7)

# Victimes de vols ou tentatives de vols sans violences ni menaces
plot_trend(df, 8)

# Victimes de vols ou tentatives de vols avec violences ou menaces
plot_trend(df, 9)

plot_trend(df, 10)

plot_trend(df, 11)

plot_trend(df, 12)

plot_trend(df, 13)

plot_trend(df, 14)

plot_trend(df, 15)

plot_trend(df, 16)

plot_trend(df, 17)

plot_trend(df, 18)


```

```python
plt.figure(figsize=(12, 6))
for row_index in [0, 1, 3]:  # Exemple : Cambriolages, vols sans effraction, vols de voiture
    plt.plot(df.columns[1:], df.iloc[row_index, 1:], marker='o', label=df.iloc[row_index, 0])
plt.xlabel('Année')
plt.ylabel('Nombre')
plt.title('Comparaison des Tendances de Différents Indicateurs')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```
